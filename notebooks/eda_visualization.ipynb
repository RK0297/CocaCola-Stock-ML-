```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data_path = os.path.join("..", "data", "coca_cola_stock_engineered.csv")
data = pd.read_csv(data_path)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
print("üîç Summary Statistics:\n")
print(data.describe())
```

```python
# 5.2 üìà Line Chart: Close + MA_20 + MA_50
plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['Close'], label='Close', linewidth=2)
plt.plot(data['Date'], data['MA_20'], label='MA 20', linestyle='--')
plt.plot(data['Date'], data['MA_50'], label='MA 50', linestyle='--')
plt.title("Coca-Cola Closing Price with 20 & 50-Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 5.3 üî• Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Coca-Cola Stock Features")
plt.tight_layout()
plt.show()
```
